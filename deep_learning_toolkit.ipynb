The following code provides basic operations for deep learning implementation
Include activation, sigmoid, derivative of sigmoid, softmax, derivative of softmax, back propagation weights update, euclidean distances calculation, self organizing map weights calculation, cost function and TLU implementation.

Usages are listed below examples.

The code provide basic understanding of most of deep learning implementation.


```python
from math import exp
from scipy import signal
import numpy as np

def mirror(arr):
  arr = arr[::-1]
  for i in range(len(arr)):
    arr[i] = arr[i][::-1]
  return np.array(arr)

def activate(inputs, weights):
  print(' input dot weights ')
  activation = 0
  for i in range(len(weights)):
    activation += weights[i] * inputs[i]
    print(weights[i], '*', inputs[i], '+')
  print('== ', activation)
  return activation
  
def sigmoid(activation):
  print(sigmoid)
  print('     1  ')
  print('-----------')
  print('1+e^-', activation)
  print('== ',1.0 / (1.0 + exp(-activation)))
 
def delta_sigmoid_out(expected, output):
  print('(t-a)*a*(1-a)')
  print('(', expected, '-', output, ')*', output, '*(1-', output, ')')
  print(' == ', (expected - output) * output * (1- output))
      
def delta_sigmoid(prev_deltas, output, weights):
  sum_val = 0
  print('sum(previous delta * weight)*a(1-a)')
  print('(')
  for i in range(len(prev_deltas)):
    sum_val += prev_deltas[i] * weights[i]
    print(prev_deltas[i], ' * ', weights[i], '+')
  print(') *', output, '*(1-', output,')')
  print(' == ', output * (1 - output) * sum_val)


def softmax_out(output):
  softmax = list()
  som_max = 0
  for i in range(len(output)):
    som_max += exp(output[i])
  for i in range(len(output)):
    softmax.append(exp(output[i]) / som_max)
    print('exp^', output[i], ' / ', som_max, ' == ', exp(output[i]) / som_max)
  

def delta_softmax_out(expected, y, a):
  print('(t-y)*(y-y^2)*a*(1-a)')
  for i in range(len(expected)):
    print('(',expected[i], '-', y[i], ')*(', y[i], '-', y[i]**2,')*', a[i], '(1-', a[i], ')',' == ', (expected[i] - y[i])* (y[i] - y[i]**2) * a[i] * (1-a[i]))
 

def delta_relu(prev_deltas, output, weights):
  sum_val = 0
  for i in range(len(prev_deltas)):
    sum_val += prev_deltas[i] * weights[i]
  if output > 0: print(sum_val) 
  elif output <= 0: print(0)

def bp_update(weight, l_r, delta, output):
  print('delta weight== l_r * delta * previous output')
  print('delta weight', l_r, '*', delta, ' * ', output, ' == ', l_r * delta * output)
  print('new weight', weight, '+', l_r * delta * output, '==', weight + l_r * delta * output)
  return weight + l_r * delta * output

def euclidean(input, weights):
  result = 0
  re = list()
  for x in range(len(weights)):
    print('distance == sqrt of ')
    for i in range(len(input)):
      result += (input[i] - weights[x][i]) ** 2
      print('(', input[i], '-', weights[x][i], ')^2 +')
    re.append(result)
    print('== sqrt(', result, ')')
    result = 0
  return re
  

def som_weight_update(l_r, input, weights):
  print('delta == l_r * (input[i] - weights[i])')
  for i in range(len(input)):
    print('delta', l_r, '*(', input[i], '-', weights[i],')',' == ', l_r * (input[i] - weights[i]))
    print('original weight==', weights[i])
    print('new weight == ', (l_r * (input[i] - weights[i])) + weights[i])


def error(target, output):
  print('1/2 * sum((target[i] - output[i])^2)')
  sum = 0
  for i in range(len(output)):
    sum += (target[i]-output[i])**2
    print('(', target[i], ' - ', output[i], ')^2 + ')
  sum /= 2 # can be change to /2
  print('== ', sum)


def som(inputs, weights, l_r):
  distances = list()
  for input in inputs:
    distances.append(euclidean(input, weights))
    x = distances[0].index(min(distances[0]))
    print('     ')
    for i in range(len(weights[x])):
      print('delta weights', l_r, '*(', input[i], '-', weights[x][i],')' ' == ', l_r *(input[i] - weights[x][i]))
      print('original weights == ', weights[x][i])
      weights[x][i] += l_r *(input[i] - weights[x][i])
      print('revised weights == ', weights[x][i])
      print('      ')
    distances.clear()
    print('all the weights == ', weights)


def bp(prev_deltas, weights, output, l_r):
  new_weights = list()
  for i in range(len(prev_deltas)):
    new_weights.append(bp_update(weights[i], l_r, prev_deltas[i], output))
    print('       ')
  print('new weights == ', new_weights)


def TLU (X, target, theta, weights, l_r):
  for x, t in zip(X, target):
    print('input == ', x, 'weights == ', weights )
    activation = activate(x, weights)
    if activation >= theta:
      result = 1
      if result == t:
        print('result == ', result ,'correct')
      elif result != t:
        for i in range(len(x)): 
          print('delta weight ', l_r, '*(', t, '-', result, ')', ' * ', x[i], ' == ' , l_r * (t - result) * x[i])
          print('new weight ==', weights[i] +  l_r * (t - result) * x[i])
          weights[i] = weights[i] +  l_r * (t - result) * x[i]
      print('weights == ', weights)
      print('    ')
    else:
      result = 0
      if result == t:
        print('result == ', result ,'correct')

      elif result != t:
        for i in range(len(x)): 
          print('delta weight ', l_r, '*(', t, '-', result, ')', ' * ', x[i], ' == ' , l_r * (t - result) * x[i])
          print('new weight ==', weights[i] +  l_r * (t - result) * x[i])
          weights[i] = weights[i] +  l_r * (t - result) * x[i]
      print('weights == ', weights)
      print('   ')
  print('         ')
  print('final weights == ', weights)


def TLU_single (x, target, theta, weights, l_r):
  activation = activate(x, weights)
  if activation >= theta:
    result = 1
    if result == target:
      print('result== ', result ,'correct')
    elif result != target:
      for i in range(len(x)): 
        print('delta weight== ', l_r * (target - result) * x[i])
        print('new weight ==', weights[i] +  l_r * (target - result) * x[i])
  else:
    result = 0
    if result == target:
      print('result== ', result ,'correct')
    elif result != target:
      for i in range(len(x)): 
        print('delta weight== ', l_r * (target - result) * x[i])
        print('new weight ==', weights[i] +  l_r * (target - result) * x[i])

  
```
